```python
# Standard imports for pandas and sklearn sample datasets
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the iris data
data = pd.read_csv("agaricus-lepiota.csv")

# Get first five values of the dataset
data.head()
```

```python
# Seperate data into edible and non-edible mushrooms
edible = data[data['edibility'] == 'p']
non_edible = data[data['edibility'] == 'e']

# Plot histograms for the 'habitat' column for edible and non-edible mushrooms
plt.hist(edible['habitat'], bins=20, alpha=0.5, label='Edible', color='blue')
plt.hist(non_edible['habitat'], bins=20, alpha=0.5, label='Non-Edible', color='orange')

# Add Labels to differenciate the graphs
plt.xlabel('Habitat')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.show()
```

```python
# Replace letters with unique dummy values which are either 0.0 or 1.0
dummies = pd.get_dummies(data, prefix=None,prefix_sep='_')

#Display the data set with the dummy values
dummies
```

```python
# Most effective way of showing data I've found (Source: Trust me bro / Kokeboken)
plt.spy(dummies, markersize=0.1)
fig = plt.gcf()
fig.set_size_inches(60,160)
plt.plot()
plt.show()
```
